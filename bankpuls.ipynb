```python
assets = assets.replace('-',0)
liab = liab.replace('-',0)
obs = obs.replace('-',0)
```

```python
BS = pd.merge(assets, liab, on=['Banks','Year'], how='left')
data = pd.merge(BS, obs, on=['Banks','Year'], how='left')
data
```

```python
data.columns
```

```python
data['liq_assets'] = data['1.     Cash in hand'] + data['2.     Balances with RBI'] +  data['3.     Balances with banks in India'] + data['4.     Money at call and short notice'] + data['5.     Balances with banks outside India']+ data['6.1.       Investments in India'] + data['6.2.       Investments outside India'] - data['(ii)    Subsidiaries and/or joint ventures'] -data['(v)    Subsidiaries and/or joint ventures'] 

data['illiq_assets'] = data['(i)         Priority sectors '] + data['(ii)    Subsidiaries and/or joint ventures'] + data['(v)    Subsidiaries and/or joint ventures'] + data['(iv)       others'] + data['7C.2.    Advances outside India']+data['8.     Fixed Assets'] + data['9.     Other Assets']

data['semiliq_assets'] = data['(ii)        Public sectors'] + data['(iii)       Banks']
```

```python
data['liq_liab'] = data['1.     Capital'] + data['2.     Reserves and Surplus'] + data['(iii)      From other institutions and agencies'] + data['4.2.      Borrowings outside India'] + data['5.     Other liabilities & provisions'] - data['5.1.      Bills Payable']

data['illiq_liab'] = data['5.1.      Bills Payable'] + data['(ii)         From others'] + data['3A.2.   Savings bank deposits']

data['semiliq_liab'] = data['(i)          From banks'] + data['3A.3.   Term deposits'] + data['(i)        From Reserve Bank of India'] + data['(ii)       From other banks']
```

```python
data['LC_catnonfat'] = 0.5*data['liq_liab'] + 0*data['semiliq_liab'] - 0.5*data['illiq_liab'] + 0.5*data['illiq_assets'] + 0*data['semiliq_assets'] -0.5*data['liq_assets']
```

```python
data['semiliq_obs'] = data['Liability on account of outstanding forward exchange contracts']

data['illiq_obs'] = data['Total'] - data['Liability on account of outstanding forward exchange contracts']
```

```python
data['LC_catfat'] = 0.5*data['liq_liab'] + 0*data['semiliq_liab'] - 0.5*data['illiq_liab'] + 0.5*data['illiq_assets'] + 0*data['semiliq_assets'] -0.5*data['liq_assets'] + 0*data['semiliq_obs'] + 0.5*data['illiq_obs'] 
```

```python
mat_profile = pd.read_excel("Maturity Profile.xlsx")
mat_profile
```

```python
mat_profile.columns
```

```python
mat_profile = mat_profile.replace('-',0)
```

```python
data['illiq_assets_mat'] = data['8.     Fixed Assets'] + data['9.     Other Assets'] + data['(ii)    Subsidiaries and/or joint ventures'] + data['(v)    Subsidiaries and/or joint ventures'] + mat_profile['Loans and Advances f)   Over 1 year to 3 years'] + mat_profile['Loans and Advances g)  Over 3 years to 5 years'] + mat_profile['Loans and Advances h)  Over 5 years']

data['semiliq_assets_mat'] = mat_profile['Loans and Advances a)  1 - 14 days'] + mat_profile['Loans and Advances b)  15 - 28 days'] + mat_profile['Loans and Advances c)  29 days to 3 months'] + mat_profile['Loans and Advances d)  Over 3 months to 6 months'] + mat_profile['Loans and Advances e)  Over 6 months to 1 year']
```

```python
data['LC_catnonfat'] = 0.5*data['liq_liab'] + 0*data['semiliq_liab'] - 0.5*data['illiq_liab'] + 0.5*data['illiq_assets_mat'] + 0*data['semiliq_assets_mat'] -0.5*data['liq_assets']

data['LC_matfat'] = 0.5*data['liq_liab'] + 0*data['semiliq_liab'] - 0.5*data['illiq_liab'] + 0.5*data['illiq_assets_mat'] + 0*data['semiliq_assets_mat'] -0.5*data['liq_assets'] + 0*data['semiliq_obs'] + 0.5*data['illiq_obs'] 
```

```python
earnings = pd.read_excel("Earnings and Expenses.xlsx", sheet_name = 'EARNINGS')
expenses = pd.read_excel("Earnings and Expenses.xlsx", sheet_name = 'EXPENSES')
pnl = pd.merge(earnings, expenses, on=['Banks','Year'], how='left')
data1 =  pd.merge(data, pnl, on=['Banks','Year'], how='left')
data1
```

```python
data1['ROA'] = np.nan
data1 = data1.replace('-',0)
data1['ROA'] = data1['VIII.  Profit (Loss) during the year']/data1['Total Assets']
```

```python
data1 = data1.sort_values(by=['Banks', 'Year'], ascending=[True, True])
data1['ROA_std'] = data1.groupby('Banks')['ROA'].rolling(window=3).std().reset_index(level=0, drop=True)
data1['deposit_growth_rate'] = data1.groupby('Banks')['3.     Deposits'].pct_change() * 100
data1 = data1.sort_index()
```

```python
data1['Z-score'] = np.log((data1['ROA'] + (data1['1.     Capital'] + data1['2.     Reserves and Surplus'])/data1['Total Assets'])/data1['ROA_std'])
```
